```python
%matplotlib inline

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os
from imutils.object_detection import non_max_suppression
import glob
```

#### Carga de imágenes y templates

```python
# Carga de todas las imágenes y el template
# Ruta a las imágenes
image_dir = './images'

# Filtramos solo archivos de imagen válidos
valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')
image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(valid_extensions)]

# Cargamos todas las imágenes en listas (una en color y otra en escala de grises)
images_rgb = []
images_gray = []
image_names = []
images_canny = []

for filename in image_files:
    path = os.path.join(image_dir, filename)
    img = cv.imread(path)
    if img is not None:
        img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        img_canny = cv.Canny(img_gray, 100, 180, L2gradient=True)
        images_rgb.append(img_rgb)
        images_gray.append(img_gray)
        images_canny.append(img_canny)
        image_names.append(filename)
    else:
        print(f'No se pudo cargar: {filename}')

# Cargamos el template en escala de grises
template = cv.imread('./template/pattern.png', 0)
template_canny = cv.Canny(template, 150, 200, L2gradient=True)

# Visualizamos el template
plt.figure(figsize=(15, 5))
plt.subplot(1, 5, 1)
plt.imshow(template, cmap='gray')
plt.title('Template')
plt.subplot(1, 5, 2)
plt.imshow(template_canny, cmap='gray')
plt.title('Template Canny')

# Visualizamos la primera imagen como ejemplo

if images_rgb:
    plt.subplot(1, 5, 3)
    plt.imshow(images_rgb[0])
    plt.title(image_names[0])
    plt.subplot(1, 5, 4)
    plt.imshow(images_gray[0], cmap='gray')
    plt.title(image_names[0] + ' B&W')
    plt.subplot(1, 5, 5)
    plt.imshow(images_canny[0], cmap='gray')
    plt.title(image_names[0] + ' Canny')
else:
    print("No se cargaron imágenes correctamente.")

plt.tight_layout()
plt.show()
```

       el método de comparación y un umbral para considerar una coincidencia válida.

       param draw: Si es True, dibuja las coincidencias encontradas en la imagen

       param default_scale: Si es True, usa escalas predefinidas, si es False, usa el rango y paso definidos (puse esto porque las fotos son muy dispares y encontre valores random que van bien)
       return: Lista de coordenadas (x, y) de las coincidencias encontradas """

```python
import numpy as np
import cv2 as cv
import imutils
import matplotlib.pyplot as plt
from imutils.object_detection import non_max_suppression
import glob
```

```python
# Función para leer las imágenes
def read_img(filename):
    img = cv.imread(filename)
    img_rgb= cv.cvtColor(img, cv.COLOR_BGR2RGB)
    #img_gray= cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img_gray = cv.imread(filename,0)
    return img_rgb, img_gray
```

```python
def template_match_single_detection(img_rgb, img_gray, template, scale_range=(0.5, 3.0), scale_step=0.1, manual_scale=False, method=cv.TM_CCOEFF_NORMED):
    if not manual_scale:
        scale = [ 0.3, 0.75, 1.5, 1.75, 2.5, 4]
    else:
        scale = np.arange(*scale_range, scale_step)
    w, h = template.shape[::-1]    


    best_score = -10
    best_res = []

    for s in range(len(scale)):
        image_resized = cv.resize(img_gray, None, fx=scale[s], fy=scale[s], interpolation=cv.INTER_AREA)
        if image_resized.shape[0] > h and image_resized.shape[1] > w:
            image_canny = cv.Canny(image_resized, 100, 180, L2gradient=True)
            res = cv.matchTemplate(image_canny, template, method)
       #     normalized_res = res
       #     normalized_res = cv.normalize(res,res,0,1, cv.NORM_MINMAX)
       #     min_val, max_val, min_loc, max_loc = cv.minMaxLoc(normalized_res) # Busco maximos
            min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
            if max_val > best_score: # si el score es mejor, guardo la posición
                best_score = max_val
                best_scale = scale[s]
                r = img_gray.shape[1] / float(image_resized.shape[1])
                best_res = res
                top_left = (int(max_loc[0] * r), int(max_loc[1] * r)) 
                bottom_right = (int((max_loc[0] + w) * r), int((max_loc[1] + h) * r))
        else: continue # Si la imagen es menor que el template, continuar con el loop

    #copiamos las imágenes para dibujar los rectángulos
    img_rgb_copy = img_rgb.copy()
    img_res_copy = image_canny.copy()
    
    
    cv.rectangle(img_rgb_copy, top_left, bottom_right,(0, 255, 255), 3)
    
    print(f"Imagen: {filename}")
    print(f"Máximo: {max_val}")
    print(f"Nivel de Confianza:  {np.round(1- best_score,3)}")
    print(f"Escala: {best_scale}")

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1),
    plt.imshow(template, cmap='gray')
    plt.title('Template con bordes de Canny')

    plt.subplot(1, 3, 2),
    plt.imshow(img_res_copy, cmap = 'gray')
    plt.title('Visualize Edged')

    # Imagen original con recuadros
    plt.subplot(1 ,3, 3),
    plt.imshow(img_rgb_copy)
    plt.title(label= f"Nivel de confianza:{np.round(1- best_score,3)}", loc='center')
    plt.show()
    return 
```
